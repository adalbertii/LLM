```python
import pandas as pd

# Odczytanie pliku CSV
plik_wejsciowy = 'miejscowosci.csv'
df = pd.read_csv(plik_wejsciowy)

# Sprawdzenie, czy istnieje kolumna 'nazwa'
if 'NAZWA' in df.columns:
    # Utworzenie listy niepowtarzalnych miejscowości
    unikalne_miejscowosci = df['NAZWA'].unique()

    # Konwersja do DataFrame
    df_wyniki = pd.DataFrame(unikalne_miejscowosci, columns=['nazwa'])

    # Zapisanie do pliku 'wyniki.csv'
    df_wyniki.to_csv('wyniki.csv', index=False)
    print("Zapisano niepowtarzalne miejscowości do pliku wyniki.csv")
else:
    print("Brak kolumny 'nazwa' w pliku wejściowym.")
```
